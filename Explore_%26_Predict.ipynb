# Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.cross_validation import train_test_split
```

```python
#For Android
y_android=pd.DataFrame(json_dataframe.android_total_ratings)

x1=json_dataframe.android_avg_rating
x2=json_dataframe.android_file_size
x3=json_dataframe.android_rating_1
x4=json_dataframe.android_rating_2
x5=json_dataframe.android_rating_3
x6=json_dataframe.android_rating_4
x7=json_dataframe.android_rating_5

#print type(x1)
#For iOS
y_ios=pd.DataFrame(json_dataframe.ios_all_ratings)

x8=json_dataframe.ios_current_ratings
x9=json_dataframe.ios_file_size


#print type(json_dataframe.android_file_size)
#print (x)
```

```python
x_android_linear=pd.concat([x1,x3,x4,x7],axis=1)
x_ios_linear=pd.concat([x1,x2,x4,x5],axis=1)
```

```python
linear_android=linear_model.LinearRegression()
linear_ios=linear_model.LinearRegression()
```

```python
#Linear regression for Android

xa_train, xa_test, ya_train,  ya_test = train_test_split(x_android_linear,y_android, test_size=0.5, random_state=4)
linear_android.fit(xa_train, ya_train)
```

```python
#Linear regression for ios

xi_train, xi_test, yi_train,  yi_test = train_test_split(x_ios_linear,y_ios, test_size=0.5, random_state=4)
linear_ios.fit(xi_train, yi_train)
#print linear_ios.coef_
```

```python
print "Linear Regression Models\n"
#print "fgnm",reg.coef_[0][0],"fd",reg.intercept_[0]
print "linear_model_android=",linear_android.intercept_[0],"+",linear_android.coef_[0][0],"* android_avg_ratings +",linear_android.coef_[0][1],"* android_rating_1",linear_android.coef_[0][2],"* android_rating_2",linear_android.coef_[0][3],"* android_rating_5\n"

print "linear_model_ios=",linear_ios.intercept_[0],"+",linear_ios.coef_[0][0],"* android_avg_ratings +",linear_ios.coef_[0][1],"* android_file_size +",linear_ios.coef_[0][2],"* android_rating_2 +",linear_ios.coef_[0][3],"* android_rating_3"
```

```python
android_lin_pred=linear_android.predict(xa_test)
ios_lin_pred=linear_ios.predict(xi_test)
#print "android pred        iOS pred"
#for i in range(len(ya_test)):
#    print android_pred[i],ios_pred[i]
```

```python
#Mean Square Error
MSE_lin_android=np.mean((android_lin_pred-ya_test)**2)/len(android_lin_pred)
MSE_lin_ios=np.mean((ios_lin_pred-yi_test)**2)/len(ios_lin_pred)
print "Mean Square Errors: For Linear Model"
print "\nMSE of Android= ", MSE_lin_android[0],'\n'
print "MSE of iOS= ", MSE_lin_ios[0],'\n'
```

```python
#R-squared value
from sklearn.metrics import r2_score

print "R2-Values: For Linear Model"

COD_android_lin = r2_score(ya_test, android_lin_pred)
print "\nR2-Value for Android= ",COD_android_lin,'\n'

COD_ios_lin = r2_score(yi_test, ios_lin_pred)
print "R2-Value for iOS= ",COD_ios_lin,'\n'
```

# Forecasting Xs


```python
#https://analysights.wordpress.com/tag/smoothing-constant/
#Forecasting of X values using Smoothing Constant Method for prediction of Y
def forecast(actual):
    #actual=act
    req=[]
    for i in range(1,720,24):
        req.append(actual[-i])
    forecasted=[req[0]]
    for i in range(len(req)):
        forecasted.append(0.8*req[i]+0.2*forecasted[i])
    return forecasted

x1_p=json_dataframe.android_avg_rating
x2_p=json_dataframe.android_file_size
x3_p=json_dataframe.android_rating_1
x4_p=json_dataframe.android_rating_2
x5_p=json_dataframe.android_rating_3
x6_p=json_dataframe.android_rating_4
x7_p=json_dataframe.android_rating_5
x8_p=json_dataframe.ios_current_ratings
x9_p=json_dataframe.ios_file_size


#Forecasted X values
x1_pred=forecast(x1_p)
x2_pred=forecast(x2_p)
x3_pred=forecast(x3_p)
x4_pred=forecast(x4_p)
x5_pred=forecast(x5_p)
x6_pred=forecast(x6_p)
x7_pred=forecast(x7_p)
x8_pred=forecast(x8_p)
x9_pred=forecast(x9_p)
#print x1_pred,x2_pred,x3_pred,x4_pred,x5_pred,x6_pred,x7_pred,x8_pred,x9_pred
```
