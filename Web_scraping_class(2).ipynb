```python
class Web_scraping:
    """returns the ios and android data values"""
    
    def __init__(self, doc1, doc2):
        self.get_ios_data_list(doc1)
        self.get_android_data_list(doc2)
    
    def get_ios_data_list(self, doc1):
        r_ios = urllib.urlopen(doc1).read()
        #print type(r)
        soup_ios = BeautifulSoup(r_ios, "lxml")
        #print soup_ios.prettify()



        ios_doc_ratings = soup_ios.find("div", {"class": "extra-list customer-ratings"}) # find first 
        #print('HTML source')
        #print('_'*30)
        #print(ios_doc_ratings)
        #print('_'*30)
        ios_rating_corpus = ios_doc_ratings.find_all("div")
        #print ios_rating_corpus#-beautifulSoup
        ios_rating_str = []
        for item in ios_rating_corpus:
            ios_rating_str.append((unicodedata.normalize('NFKD', (item.get_text())).encode('ascii','ignore').split()))
        #print ios_rating_str
        ios_data_numbers = []
        for x in ios_rating_str:
            for j in range(len(x)):
                if x[j]=='Ratings':
                    ios_data_numbers.append(eval(x[j-1]))
        #ios_current_rating = ios_rating_numbers[0]
        #ios_all_rating = ios_rating_numbers[1]
        #print ios_data_numbers
        #print(ahrefs)
        ios_doc_size = soup_ios.find("ul", {"class": "list"}) # find first 
        #print('HTML source')
        #print('_'*30)
        #print(ios_doc_size)
        #print('_'*30)
        ios_size_corpus = ios_doc_size.find_all("li")
        #print type(ios_size_corpus)-<class 'bs4.element.ResultSet'>
        #print ios_size_corpus
        ios_size_str = []

        for item in ios_size_corpus:
            ios_size_str.append((unicodedata.normalize('NFKD', (item.get_text())).encode('ascii','ignore').split()))
        #print ios_size_str
        #ios_size_numbers = []
        for x in ios_size_str:
            for j in range(len(x)):
                if x[j]=='Size:':
                    ios_data_numbers.append(eval(x[j+1]))
        #ios_file_size = ios_size_numbers[0]
        #ios_all_rating = ios_rating_numbers[1]
        return ios_data_numbers#, ios_all_rating
        #print(ahrefs)            
        
    def get_android_data_list(self, doc2):
        r_android = urllib.urlopen(doc2).read()
        #print type(r)
        soup_android = BeautifulSoup(r_android, "lxml")
        android_data_numbers = []
        

        android_doc_ratings = soup_android.find("div", {"class": "score-container"}) # find first 
        #print('HTML source')
        #print('_'*30)
        #print(android_doc_ratings)
        #print('_'*30)
        #import unicodedata
        #print type(android_doc_ratings),android_doc_ratings
        if android_doc_ratings is not None:

            android_rating_corpus = android_doc_ratings.find_all("div")
            #print type(ios_rating_corpus)-beautifulSoup
            android_rating_str = []
            for item in android_rating_corpus:
                android_rating_str.append((unicodedata.normalize('NFKD', (item.get_text())).encode('ascii','ignore').split()))
            #print android_rating_str
            
            for i in range(len(android_rating_str)):
                if i==0:
                    android_data_numbers.append(android_rating_str[0][0])
                for j in range(len(android_rating_str[i])):
                    if (android_rating_str[i])[j]=='total':
                        android_data_numbers.append(eval((android_rating_str[i][j-1]).replace(',','')))
            #for i in android_data_numbers:
                #print type(i), i 
            #android_current_rating = android_rating_numbers[0]
            #android_all_rating = android_rating_numbers[1]
            #print android_rating_numbers#, android_current_rating, android_all_rating
            #print(ahrefs)'''

            android_doc_ratings = soup_android.find("div", {"class": "rating-histogram"}) # find first 
            #print('HTML source')
            #print('_'*30)
            #print(android_doc_ratings)
            #print('_'*30)

            #getting ratings

            #http://stackoverflow.com/questions/1207457/convert-a-unicode-string-to-a-string-in-python-containing-extra-symbols
            #import unicodedata

            android_rating_corpus = android_doc_ratings.find_all("div")
            #print type(ios_rating_corpus)-beautifulSoup
            android_rating_str = []
            for item in android_rating_corpus:
                android_rating_str.append((unicodedata.normalize('NFKD', (item.get_text())).encode('ascii','ignore').split()))
            #print android_rating_str
            #android_rating_numbers = []
            for i in reversed(android_rating_str):
                #print i
                for j in range(len(i)):
                    if j%2==0:
                        android_data_numbers.append(eval(i[1].replace(',','')))
        #android_current_rating = android_rating_numbers[0]
        #android_all_rating = android_rating_numbers[1]
        #return android_data_numbers
        #for i in range(len(android_rating_numbers)):
        #    android_rating_numbers[i]=eval(android_rating_numbers[i].replace(',',''))#, android_current_rating, android_all_rating
        #print(ahrefs)
        #for i in android_data_numbers:
        #    print type(i), i
        android_doc_size = soup_android.find("div", {"class": "content", "itemprop": "fileSize"}) # find first 
        #print('HTML source')
        #print('_'*30)
        #print(android_doc_size)
        #print('_'*30)
        #getting Size

        #import unicodedata

        android_size_corpus = str(android_doc_size).split()#.find()
        #print type(ios_size_corpus)-<class 'bs4.element.ResultSet'>
        #print android_size_corpus
        #android_size_str = []

        #for item in android_size_corpus:
        #    android_size_str.append((unicodedata.normalize('NFKD', (item.get_text())).encode('ascii','ignore').split()))
        #print ios_size_corpus
        android_size_numbers = []
        for i in range(len(android_size_corpus)):
            if android_size_corpus[i]=='</div>':
                android_size_numbers = android_size_corpus[i-1].replace('M','')
        #ios_file_size = ios_size_numbers[0]
        #ios_all_rating = ios_rating_numbers[1]
        if len(android_size_numbers)!=0:
            android_data_numbers.append((android_size_numbers))

        #for i in range(len(android_rating_numbers)):
        #    if android_rating_numbers[i] == str():
        #        android_rating_numbers[i]=float(android_rating_numbers[i].replace(',',''))
        return android_data_numbers
        #print(ahrefs)'''
        
```
